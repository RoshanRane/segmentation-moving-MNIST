## Demonstration of how to 
1. Generate the dataset with 2 to 4 MNIST digits per video.
2. Display some sample videos from the dataset.
3. Save the generated dataset as '.h5'  file.

```python
from moving_mnist_for_segmentation.dataset import Dataset 
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
# set random seeds for replicability
import numpy as np
import random
np.random.seed(42)
random.seed(42)
```

```python
dataset = Dataset()
dat, dat_masks, dat_labels = dataset.generate(
        split="val",
        frame_size=(64, 64),
        num_videos=100, 
        num_frames=20, 
        min_ndigits=2, max_ndigits=4,
        ignore_label=10)

print("Videos: {} \nSegmentation Masks: {} \nMNIST Labels: {}".format(
    dat.shape, dat_masks.shape, dat_labels.shape))
```

```python
dataset.display_samples(4)
```

```python
dataset.save(path = "data/", filetype = "h5")
```
